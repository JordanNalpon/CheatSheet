#### <font color = yellow> Standard Deviation </font>

<img src="images/StandDev.png">

```python
#standard deviation code

def stand_dev(data):
    
    # n is the number of values in the column
    n = len(data)
    # mean is the total sum of values in the column divided by n
    mean = (data.sum(axis = 0)) / n
    
    # x is used as a placeholder value to store the sum of ( (current value - mean)**2 / n )
    x = 0
    
    # the value i will loop through all the values in the inputted column and generate a value for dev. The dev value will be added to x which will store the value while the loop repeats with a different i value.
    for i in data:
        dev = (((i - mean)**2)/n)
        x += dev
    
    # the final value of x needs to be squared to get our standard deviation
    return x**0.5
```

```python
#Below we will see what is the stand dev for all the columns
sd = {col :stand_dev(df_train[col]) for col in df_train.iloc[: , 1 : :].columns}

sd
```
